```python
# btm as dataframe
btm_plus = pd.DataFrame.from_dict(btm)
btm_plus
```

```python
# export as .csv
btm_plus.to_csv('datasets/BTM_plus_original.csv', sep='\t', index=False, header=False)
```

```python
# find the longest list of genes
list = []
for i in range(btm_plus['genes'].size):
    list.append(len(btm_plus['genes'][i]))
longest = max(list)
longest
```

```python
df = btm_plus
df['pathway'] = df['name'] + ' (' + df['id'] + ')'
df
```

```python
# create more 98 empty colums
df = pd.concat([df,pd.DataFrame(columns=range(longest))])
#df = df.where(pd.notnull(df), '')
df
```

```python
# fill the colums with the genes
for i in range(df.shape[0]):
    for j in range(len(df['genes'][i])):
        df.loc[i, j] = df['genes'][i][j]
df
```

```python
# drop not required columns
df = df.drop(['id', 'name', 'src', 'genes'], axis=1)
df
```

```python
# export as .gmt
df.to_csv('datasets/BTM_plus_for_GSEA.gmt', sep='\t', index=False, header=False)
```
